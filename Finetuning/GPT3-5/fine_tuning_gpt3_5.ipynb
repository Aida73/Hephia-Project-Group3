```python
from openai import OpenAI

client = OpenAI(
    # defaults to os.environ.get("OPENAI_API_KEY")
    api_key = '',
)
```

```python
def predict(test_messages, fine_tuned_model_id):
    response = client.chat.completions.create(
        model=fine_tuned_model_id,
            messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant designed to output JSON.",
        },
        {"role": "user", "content": test_messages},
    ],
        temperature=1,
        max_tokens=1000
    )
    #return response.content
    return response.choices[0].message.content
     #   response_format={"type": "json_object"},
```

```python
def append_result_to_csv(result, csv_path):
    # Créer le fichier CSV s'il n'existe pas encore
    if not os.path.isfile(csv_path):
        result_df = pd.DataFrame(columns=result.keys())
        result_df.to_csv(csv_path, index=False)

    # Ajouter le résultat au fichier CSV existant
    result_df = pd.DataFrame(result, index=[0])
    result_df.to_csv(csv_path, mode='a', header=not pd.read_csv(csv_path).shape[0], index=False)

# Charger le fichier CSV existant
csv_file_path = "results_gpt_ft(2).csv"

# Charger les données du fichier CSV dans un DataFrame
df = pd.read_csv("data (1).csv")

# Boucle sur chaque question (exécutez cette cellule pour chaque question)
question_index = 9  # Remplacez par l'index de votre question dans le DataFrame

# Extraire la question et la réponse attendue
question = df.loc[question_index, 'Question']
# expected_response = df.loc[question_index, 'Answers']

# Mesurer le temps d'exécution
start_time = time.time()

# Générer la réponse avec GPT-3.5 finetuned
response = predict(question, fine_tuned_model_id)


# Mesurer le temps d'exécution
end_time = time.time()
execution_time = end_time - start_time

# Créer un dictionnaire avec les résultats
result_dict = {
    'Question': question,
    'Réponses GPT-3.5 finetuned': response,
    'Temps d\'exécution': execution_time
}

# Ajouter le résultat au fichier CSV existant
append_result_to_csv(result_dict, csv_file_path)

print("Résultat ajouté au fichier CSV.")
```
